```python
# кількість проданих смартфонів моделі A протягом тижня

import statistics


a_sales = [16, 25, 18, 19, 22, 34, 20]

mean_a_sales = sum(a_sales) / len(a_sales)
median_a_sales = statistics.median(a_sales)

print(f"mean_a_sales: {mean_a_sales}\nmedian_a_sales: {median_a_sales}")
```

```python
# кількість проданих смартфонів моделі B протягом тижня

import statistics


b_sales = [23, 20, 36, 12, 35, 54, 36]

mean_b_sales = statistics.mean(b_sales)
median_b_sales = statistics.median(b_sales)

print(f"mean_b_sales: {round(mean_b_sales, 2)}\nmedian_b_sales: {median_b_sales}")
```

**Аналіх кількісті проданих смартфонів моделі A протягом тижня**

Медіана приблизно дорівнює середньому, тобто вибірка симетрична.
Набір даних можна назвати збалансованим.

**Аналіз кількісті проданих смартфонів моделі B протягом тижня**

Медіана більше середнього, це вказує на те, що у вибірці переважна більшість значень великі (більші за середнє)

```python
from statistics import mode


sales = ["Samsung", "LG", "Sony", "Samsung", "LG", "Sony", "Samsung", "LG", "Sony", "Samsung"]
bestseller = mode(sales)

print(f"The bestseller is {bestseller}")
```
