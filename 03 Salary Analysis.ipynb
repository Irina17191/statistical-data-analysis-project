```python
# 2. Статистичний метод визначення викидів (Statistical Methods)
# Метод міжквартильного діапазону (IQR Method)

import numpy as np


salaries = [3800, 3100, 3700, 4000, 4000, 3700, 2400, 1900, 3600, 4600,
4800, 4500, 2200, 3700, 2900, 10500, 2700, 4600, 3600, 3400,
12000, 1800, 300, 4000, 3200, 1100, 4300, 5000, 4300, 4300,
4300, 3200, 3700, 7000, 2400, 2300, 3900, 3400, 3900, 4800,
3000, 2600, 4700, 4200, 4100, 3600, 4000, 4500, 3500, 4600]


# Calculating quartiles
Q1 = np.percentile(salaries, 25)
Q3 = np.percentile(salaries, 75)

# Interquartile range
IQR = Q3 - Q1

# Defining the boundaries for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detecting outliers
outliers = [x for x in salaries if x < lower_bound or x > upper_bound]
print("Outliers:", outliers) # Outliers: [10500, 12000, 300, 1100, 7000]

```

```python
# Z-score Method (Метод Z-оцінки) для виявлення викидів

from scipy import stats


salaries = [3800, 3100, 3700, 4000, 4000, 3700, 2400, 1900, 3600, 4600,
4800, 4500, 2200, 3700, 2900, 10500, 2700, 4600, 3600, 3400,
12000, 1800, 300, 4000, 3200, 1100, 4300, 5000, 4300, 4300,
4300, 3200, 3700, 7000, 2400, 2300, 3900, 3400, 3900, 4800,
3000, 2600, 4700, 4200, 4100, 3600, 4000, 4500, 3500, 4600]


# Calculating Z-scores for the sales data
z_scores = stats.zscore(salaries)

# Identifying outliers based on Z-scores
outliers_z = [salaries[i] for i in range(len(salaries)) if abs(z_scores[i]) > 3]
print("Outliers based on Z-scores:", outliers_z) # Outliers based on Z-scores: [10500, 12000]
```

```python
# Гістограма для виявлення розподілу заробітної плати

import numpy as np
import matplotlib.pyplot as plt


salaries = [3800, 3100, 3700, 4000, 4000, 3700, 2400, 1900, 3600, 4600,
4800, 4500, 2200, 3700, 2900, 10500, 2700, 4600, 3600, 3400,
12000, 1800, 300, 4000, 3200, 1100, 4300, 5000, 4300, 4300,
4300, 3200, 3700, 7000, 2400, 2300, 3900, 3400, 3900, 4800,
3000, 2600, 4700, 4200, 4100, 3600, 4000, 4500, 3500, 4600]     # Outliers: [10500, 12000, 300, 1100, 7000]

# Creating a histogram to visualize the distribution
plt.hist(salaries, bins=10, color="blue", alpha=0.7)
plt.title("Distribution of Salaries")
plt.xlabel("Salaries")
plt.ylabel("Frequency")
plt.show()
```

Зображена гистограма розподілу заробітних плат схожа на правосторонній (позитивно асиметричний) розподіл. Це типова картина для даних, у яких є деякі високі значення (викиди), що "тягнуть" розподіл вправо, створюючи більшу частину маси в лівій частині.

Гістограма показує, що більшість зарплат знаходяться в діапазоні від 2000 до 6000, з кількома викидами, які підвищують частоту в правій частині (наприклад, 10500, 12000). Ці викиди можуть свідчити про наявність кількох високих значень, що впливають на загальний вигляд розподілу.

Викиди (outliers): є кілька значень, які значно більші за інші (наприклад, зарплати 10500 і 12000). Вони можуть бути результатом нетипових ситуацій або винятків, і вони сильно впливають на форму розподілу, створюючи "хвіст" праворуч. Це характерно для правосторонньої асиметрії.

Такий розподіл часто зустрічається в реальних даних, де багато значень сконцентровані в межах певного діапазону, а деякі значення є значно вищими або нижчими за основну масу, що створює асиметричний розподіл.

Отже, це схоже на **Логнормальний розподіл**: Це розподіл, в якому значення не можуть бути від'ємними, і їх логарифм має нормальний розподіл. Такий розподіл часто спостерігається в даних, що містять багато малих значень і кілька великих значень (викидів), що призводить до правосторонньої асиметрії. У нашому випадку велика частина зарплат знаходиться в діапазоні від 2000 до 6000, і лише кілька викидів (наприклад, 10500, 12000) підвищують середнє значення, створюючи «хвіст» у правій частині.


Спробємо видалити викиди із набору даних та знову побудувати гістограму.
Подивимось як зміниться характер розподілу.

```python
# Гістограма для виявлення розподілу заробітної плати

import numpy as np
import matplotlib.pyplot as plt


# видаляю з даних викиди
salaries = [3800, 3100, 3700, 4000, 4000, 3700, 2400, 1900, 3600, 4600,
4800, 4500, 2200, 3700, 2900, 2700, 4600, 3600, 3400,
1800, 4000, 3200, 4300, 5000, 4300, 4300, 4300, 3200,
3700, 2400, 2300, 3900, 3400, 3900, 4800, 3000, 2600,
4700, 4200, 4100, 3600, 4000, 4500, 3500, 4600]


# Creating a histogram to visualize the distribution
plt.hist(salaries, bins=10, color="blue", alpha=0.7)
plt.title("Distribution of Salaries")
plt.xlabel("Salaries")
plt.ylabel("Frequency")
plt.show()
```

Дана гістограма, яку утворилась після видалення екстремальних значень, розподіл виглядає більш симетричним і підходить до **нормального розподілу**, оскільки:

Симетрія: Розподіл більш рівномірно розподілений навколо центральної області (приблизно 3500), з певною кількістю значень як зліва, так і справа від середнього.

Форма "дзвінка": Гістограма схожа на класичну форму для нормального розподілу, де більшість значень зосереджені в центральній частині, і ймовірність значень зменшується у міру віддалення від середнього.

Така форма вказує на те, що розподіл близький до нормального. Це означає, що значення заробітних плат розподілені навколо середнього (найбільше зустрічаються близько до певної суми), і ймовірність дуже високих або низьких значень значно зменшується.

Однак, щоб точно сказати, чи це нормальний розподіл, можна провести більш детальний статистичний аналіз, наприклад, за допомогою тестів на нормальність (наприклад, тест Шапіро-Уїлка або Колмогорова-Смірнова).
