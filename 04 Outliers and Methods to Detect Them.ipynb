```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sales data
sales_data = [20, 22, 23, 500, 19, 21, 30, 45, 18, 25, 23]

# Creating a line plot to detect outliers
sns.lineplot(sales_data, marker="o")
plt.title("Dynamics of Sales")
plt.xlabel("Days")
plt.ylabel("Sales, $")
plt.show()
```

```python
import numpy as np

# Sales data
sales_data = [20, 22, 23, 500, 19, 21, 30, 45, 18, 25, 23]

# Calculating quartiles
Q1 = np.percentile(sales_data, 25)
Q3 = np.percentile(sales_data, 75)

# Interquartile range
IQR = Q3 - Q1

# Defining the boundaries for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detecting outliers
outliers = [x for x in sales_data if x < lower_bound or x > upper_bound]
print("Outliers:", outliers) # Outliers: [500, 45]
```

```python
import numpy as np


visitors = [150, 200, 180, 220, 190, 250, 170, 1600, 210, 230]

Q1 = np.percentile(visitors, 25)
Q3 = np.percentile(visitors, 75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [outlier for outlier in visitors if outlier < lower_bound or outlier > upper_bound]
print(outliers)
```

```python
from scipy import stats

# Sales data
sales_data = [20, 22, 23, 500, 19, 21, 30, 45, 18, 25, 23]

# Calculating Z-scores for the sales data
z_scores = stats.zscore(sales_data)
print(z_scores)

# Identifying outliers based on Z-scores
outliers_z = [sales_data[i] for i in range(len(sales_data)) if abs(z_scores[i]) > 3]
print("Outliers based on Z-scores:", outliers_z) # Outliers based on Z-scores: [500]
```
