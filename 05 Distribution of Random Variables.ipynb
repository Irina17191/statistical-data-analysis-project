```python
import numpy as np
import matplotlib.pyplot as plt

# Generating data for a normal distribution
sales_data = np.random.normal(loc=50, scale=10, size=1000)

# Creating a histogram to visualize the distribution
plt.hist(sales_data, bins=20, color="blue", alpha=0.7)
plt.title("Normal Distribution of Daily Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Generating data for a uniform distribution
uniform_data = np.random.uniform(low=0, high=100, size=1000)

# Creating a histogram to visualize the uniform distribution
plt.hist(uniform_data, bins=20, color="blue", alpha=0.7)
plt.title("Uniform Distribution of Sales across Countries")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# Generating data for a binomial distribution
n, p = 10, 0.5  # 10 trials (website visits), 50% probability of success (purchase)
binom_data = binom.rvs(n=n, p=p, size=1000)

# Creating a histogram to visualize the binomial distribution
plt.hist(binom_data, bins=20, color="blue", alpha=0.7)
plt.title("Binomial Distribution of Purchase Success")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```
