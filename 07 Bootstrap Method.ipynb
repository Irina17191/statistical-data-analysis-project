**Bootstrap Method**

Маємо набір даних для інтернет-магазину з 15 транзакцій. Кількість проданого товару у цих транзакціях була такою:

 ```
 5, 15, 7, 9, 2, 10, 11, 25, 7, 8, 2, 3, 10, 10, 8
 ```

Застосуй метод bootstrap для оцінки медіани кількості проданих товарів.

2. Побудуй графік розподілу медіан для повторних вибірок. Яке значення медіани є найбільш ймовірним?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Creating a dataset of sales
data = {"Quantity": [5, 15, 7, 9, 2, 10, 11, 25, 7, 8, 2, 3, 10, 10, 8]}
quantity_data = pd.DataFrame(data)

# Bootstrap function: generating new samples with replacement
def bootstrap(data: pd.DataFrame, n_iterations: int) -> list:
    sample_medians = []
    for i in range(n_iterations):
        sample = data.sample(frac=1, replace=True)  # resampling with replacement
        sample_medians.append(sample["Quantity"].median())
    return sample_medians


# Applying bootstrap to estimate the mean
bootstrap_medians = bootstrap(quantity_data, 1000)

# Visualizing the distribution of bootstrap means
plt.hist(bootstrap_medians, bins=30, edgecolor="black")
plt.title("Bootstrap Distribution of Sample Medians")
plt.xlabel("Median Quantity")
plt.ylabel("Frequency")
plt.show()
````

Найбільш ймовірне значення медіани --- 8 одиниць.
