````python
import numpy as np
import scipy.stats as stats


# Example dataset of purchase amounts
purchase_amounts = np.array([200, 220, 230, 210, 250, 190, 215, 260, 240, 230])

# Calculating the sample mean
mean_purchase = np.mean(purchase_amounts)
print(mean_purchase)

# Calculating the standard deviation (sample-based)
std_dev = np.std(purchase_amounts, ddof=1)

# Sample size
n = len(purchase_amounts)

# Calculating the confidence interval (95%)
confidence = 0.95
confidence_interval = stats.t.interval(
    confidence, df=n - 1, loc=mean_purchase, scale=std_dev / np.sqrt(n)
)

# Printing the calculated confidence interval
print(f"Confidence Interval: {confidence_interval}")
````

1. Маємо набір даних з 15 значень про кількість проданого товару в інтернет-магазині.

 ```
 5, 15, 7, 9, 2, 10, 11, 25, 7, 8, 2, 3, 10, 10, 8
 ```

2. Використовуємо Python для обчислення 95% довірчого інтервалу для середнього значення кількості проданого товару.

3. Інтерпретуємо отриманий інтервал довіри, а саме що це означає для бізнес-рішень


````python
import numpy as np
import scipy.stats as stats


# Example dataset of purchase amounts
purchase_amounts = np.array([5, 15, 7, 9, 2, 10, 11, 25, 7, 8, 2, 3, 10, 10, 8])

# Calculating the sample mean
mean_purchase = np.mean(purchase_amounts)

# Calculating the standard deviation (sample-based)
std_dev = np.std(purchase_amounts, ddof=1)

# Sample size
n = len(purchase_amounts)

# Calculating the confidence interval (95%)
confidence = 0.95
confidence_interval = stats.t.interval(
    confidence, df=n - 1, loc=mean_purchase, scale=std_dev / np.sqrt(n)
)

# Printing the calculated confidence interval
print(f"Confidence Interval: {confidence_interval}")
````

Отже, з ймовірністю 95 % наш магазин буде продавати від 6 до 12 одиниць товару. Інтервал доволі широкий, що говорить про різноманітність значень у вибірці, а також, ймовірно, про наявність викидів. Для бізнесу це означає ненадійність оцінок середнього значення, високу ймовірність його відхилень як у позитивний, так і у негативний бік.
