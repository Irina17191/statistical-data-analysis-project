**Statistical Significance of the Results.\
Проводимо t-test та рахуємо p-value**

```python
from scipy import stats

# Sales data
without_discount = [150, 180, 160]
with_discount = [200, 220, 240]

# Perform t-test
t_stat, p_value = stats.ttest_ind(without_discount, with_discount)

# Print the p-value
print(f"p-value: {p_value}")
print(f"t_stat: {t_stat}")
```

У нашому випадку ми отримали p-value = 0.0175, а це означає, що лише 1.75 % ймовірності, що різниця у продажах між товарами зі знижкою та без знижки є випадковою.

p-value = 0.0175 < 0.05, тому ми **можемо зробити висновок, що знижка статистично значуще впливає на збільшення продажів**.
