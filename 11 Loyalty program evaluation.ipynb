```python
from scipy import stats


# Sales data
control_group = [1, 2, 1, 3, 3, 2, 1]      # контрольна група, яка не брала участь у програмі лояльності
experimental_group = [3, 4, 3, 5, 4, 4]    # експериментальна група

# Perform t-test
t_stat, p_value = stats.ttest_ind(control_group, experimental_group)

# Print the p-value
print(f"p_value: {p_value}")
```

У нашому випадку ми отримали p-value = 0.0013, а це означає, що лише 0.13 % ймовірності, що різниця у продажах між товарами зі знижкою та без знижки є випадковою.

p-value = 0.0013 < 0.05, тому ми **можемо зробити висновок, що знижка статистично значуще впливає на збільшення продажів**.
