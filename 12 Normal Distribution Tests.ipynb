```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import normaltest, shapiro

# Creating two data sets
np.random.seed(42)

# 1. Normal distribution
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# 2. Non-normal distribution (exponential)
non_normal_data = np.random.exponential(scale=2, size=1000)


# Normality tests

# Normaltest
_, p_value_normaltest_normal = normaltest(normal_data)
_, p_value_normaltest_non_normal = normaltest(non_normal_data)

# Shapiro-Wilk test
_, p_value_shapiro_normal = shapiro(normal_data)
_, p_value_shapiro_non_normal = shapiro(non_normal_data)

# Test results
print(f"Normaltest for normal data p-value: {p_value_normaltest_normal}")
print(f"Normaltest for non-normal data p-value: {p_value_normaltest_non_normal}")

print(f"Shapiro-Wilk for normal data p-value: {p_value_shapiro_normal}")
print(f"Shapiro-Wilk for non-normal data p-value: {p_value_shapiro_non_normal}")

# Plotting histograms to visualize data distributions
plt.figure(figsize=(12, 5))

# Normal distribution
plt.subplot(1, 2, 1)
plt.hist(normal_data, bins=30, color="skyblue", edgecolor="black")
plt.title("Normal Distribution")

# Non-normal distribution
plt.subplot(1, 2, 2)
plt.hist(non_normal_data, bins=30, color="salmon", edgecolor="black")
plt.title("Non-Normal (Exponential) Distribution")

plt.tight_layout()
plt.show()
```
