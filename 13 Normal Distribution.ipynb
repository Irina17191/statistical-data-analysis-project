```python
import matplotlib.pyplot as plt
from scipy.stats import normaltest, shapiro


# Given two data sets
salaries = [3000, 3200, 3500, 3700, 3100, 4000, 3300, 4500, 2900, 4200, 3400, 3100,
            2950, 3600, 3800, 5000, 3700, 5200, 4700, 5500, 4300, 6000, 3100, 2800,
            5000, 3200, 4100, 3900, 4500, 3000]

work_experience = [5.9, 4.9, 5.6, 7.3, 4.8, 4.8, 7.7, 6.6, 3.2, 6.3, 3.9, 3.9, 5.5,
                   3.9, 7.3, 6.7, 5.3, 2.9, 6.6, 5.1, 4.6, 4.6, 5.6, 8.5, 3.7, 4.4,
                   6.4, 4.5, 5.5, 3.6]

# Normality tests

# Normaltest
_, p_value_normaltest_salaries = normaltest(salaries)
_, p_value_normaltest_work_experience = normaltest(work_experience)

# Shapiro-Wilk test
_, p_value_shapiro_salaries = shapiro(salaries)
_, p_value_shapiro_work_experience = shapiro(work_experience)

# Test results
print(f"Normaltest for salaries p-value: {p_value_normaltest_salaries}")
print(f"Normaltest for work_experience p-value: {p_value_normaltest_work_experience}")

print(f"Shapiro-Wilk test for salaries p-value: {p_value_shapiro_salaries}")
print(f"Shapiro-Wilk test for work_experience p-value: {p_value_shapiro_work_experience}")



# Plotting histograms to visualize data distributions
plt.figure(figsize=(12, 5))

# salaries
plt.subplot(1, 2, 1)
plt.hist(salaries, bins=30, color="skyblue", edgecolor="black")
plt.title("Salaries Distribution")

# work_experience
plt.subplot(1, 2, 2)
plt.hist(work_experience, bins=30, color="salmon", edgecolor="black")
plt.title("Work experience Distribution")

plt.tight_layout()
plt.show()
```

**Аналіз нормальності розподілу зарплат та досвіду роботи ІТ-спеціалістів**

У межах завдання було здійснено перевірку нормальності розподілу двох вибірок: місячних зарплат ІТ-спеціалістів та їхнього професійного досвіду (у роках). Для цього застосовано два статистичних тести: тест **Д'Агостіно-Пірсона** та тест **Шапіро–Уілка**. Додатково для візуального аналізу було побудовано гістограми розподілу значень.


**Результати:**

**Тест Д'Агостіно-Пірсона:**\
для salaries p-value: 0.18071878885695186\
для work_experience p-value: 0.6172165634201328

Згідно тесту Д'Агостіно-Пірсона salaries має нормальний розподіл даних, адже p > 0.05
Також згідно тесту Д'Агостіно-Пірсона work_experience теж має нормальний розподіл даних, адже p > 0.05


**Тест Шапіро-Уілка:**\
для salaries p-value: 0.031783698516328236\
для work_experience p-value: 0.7653602896426936

**Згідно тесту Шапіро-Уілка salaries має не нормальний розподіл даних, адже p < 0.05**\
Також **згідно тесту Шапіро-Уілка work_experience має нормальний розподіл даних, адже p > 0.05**

Для невеликого розподілу даних більше підходить тест Шапіро-Уілка (функція shapiro()), тому в даному випадку будемо орієнтуватись на нього.

(Розміри обох вибірок: 30 значень, тобто менше 50 → у таких випадках рекомендовано орієнтуватися на тест Шапіро-Уілка)

normaltest() — тест Д'Агостіно-Пірсона, який працює добре при n > 50.\
shapiro() — Шапіро-Уілка, який є еталонним для малих вибірок.

**Аналіз графіків**\
Лівий графік: **Salaries Distribution**\
Візуально має асиметрію, легкий right-skew (хвіст праворуч).

Не виглядає як нормальний розподіл, що підтверджує результат тесту Шапіро.

Піки біля 3000–3200 і кілька вищих значень (5500–6000) — це може викликати відхилення від нормальності.

Правий графік: **Work experience Distribution**\
Візуально більш розсіяний, але без сильних асиметрій.

Більш наближений до нормального розподілу.

Підтверджує висновок з тесту Шапіро.

Розподіл зарплат має статистично значуще відхилення від нормального (p < 0.05), отже, гіпотеза H₀ про нормальність розподілу відхиляється.

Розподіл досвіду роботи не має статистично значущих відхилень (p > 0.05), отже, гіпотеза H₀ не відхиляється, і розподіл вважається нормальним.

Візуальний аналіз гістограм підтверджує ці результати: розподіл зарплат демонструє асиметрію з довгим правим хвостом, тоді як розподіл досвіду виглядає симетричним та близьким до нормального.

Таким чином, при подальшому аналізі даних необхідно враховувати відсутність нормальності в розподілі зарплат, що може вимагати використання **непараметричних статистичних методів**.
