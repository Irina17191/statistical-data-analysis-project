```python
import seaborn as sns

# Load the tips dataset
tips = sns.load_dataset("tips")

print(tips)
```

Крок 1. Візуалізую взаємозв'язок між розміром чайових tip та розміром чека total_bill за допомогою діаграми розсіювання.

```python
import matplotlib.pyplot as plt
import pandas as pd


# Створюємо діаграму розсіювання
plt.scatter(tips['total_bill'], tips['tip'])

# Додаємо заголовки та підписи осей
plt.title('Взаємозв\'язок між розміром чека та чайовими')
plt.xlabel('Розмір чека (total_bill)')
plt.ylabel('Розмір чайових (tip)')

# Показуємо діаграму
plt.show()

# Приклад даних
#x = [1, 2, 3, 4, 5]
#y = [2, 3, 5, 7, 11]

# Створення діаграми розсіювання
#plt.scatter(x, y)

# Додавання заголовків та підписів
#plt.title('Діаграма розсіювання')
#plt.xlabel('Ось X')
#plt.ylabel('Ось Y')

# Показ діаграми
#plt.show()
```

Крок 2. Розрахую коефіцієнти кореляції Пірсона та Спірмена між розміром чайових та розміром чека. Проаналізую чи є взаємозв'язок між ними статистично значущим.

```python
from scipy.stats import pearsonr, spearmanr


# Pearson calculations

# Calculate statistic and p-value
statistic_and_p_value_pearson = pearsonr(tips['total_bill'], tips['tip'])

# Calculate Pearson correlation
pearson_corr = tips["total_bill"].corr(tips['tip'])

print(f"Pearson Test statistic and p-value: {statistic_and_p_value_pearson}")
print(f"Pearson Correlation: {pearson_corr}")


# Spearman calculations

# Calculate statistic and p-value
statistic_and_p_value_spearman = spearmanr(tips['total_bill'], tips['tip'])

# Calculate Spearman correlation
spearman_corr = tips['total_bill'].corr(tips['tip'], method="spearman")

print(f"Spearman Test statistic and p-value: {statistic_and_p_value_spearman}")
print(f"Spearman Correlation: {spearman_corr}")
```

**Статистика:** 0.6757 — це значення кореляцій Пірсона та Спірмена, що вказує на помірно сильний позитивний лінійний взаємозв'язок між розміром чека та чайовими.

**P-value:** p-value є дуже малим, що вказує на те, що кореляція є статистично значущою (тобто ми відкидаємо нульову гіпотезу про відсутність кореляції).

Крок 3. За допомогою візуалізації визначимо, чи схожий розподіл змінних tip та total_bill на нормальний; який з двох коефіцієнтів кореляції більше підходить під ці вхідні дані.

```python
# Plotting histograms to visualize data distributions
plt.figure(figsize=(12, 5))


# tips["total_bill"]    tips['tip']
# tip data
plt.subplot(1, 2, 1)
plt.hist(tips['tip'], bins=30, color="skyblue", edgecolor="black")
plt.title("Tip data Distribution")

# total_bill data
plt.subplot(1, 2, 2)
plt.hist(tips['total_bill'], bins=30, color="salmon", edgecolor="black")
plt.title("Total bill data Distribution")

plt.tight_layout()
plt.show()
```

**Який з двох коефіцієнтів кореляції більше підходить під ці вхідні дані: Пірсона чи Спірмена?**

**Pearson:** використовується для кількісних неперервних даних, які мають нормальний розподіл.

**Spearman:** цей метод підходить для ранжованих або порядкових даних, а також для кількісних даних (у тому числі дискретних), які не є нормально розподіленими.


**Висновок:**
Гістограми показують що:

**Розподіл «tip»:**\
Явно асиметричний (правобічний), з великою кількістю малих значень і незначною кількістю великих. Немає ознак нормального розподілу.

**Розподіл «total_bill»**:
Також асиметричний, хоча й менш виражено. Значення зміщені вліво, з довгим правим «хвостом».

Оскільки обидві змінні не є нормально розподіленими, хоча й кількісні (неперервні), коефіцієнт кореляції Пірсона — не оптимальний вибір. У такій ситуації:

**Найкраще використовувати коефіцієнт Спірмена**, оскільки:
– він не вимагає нормальності розподілу,
– враховує монотонні, а не лише лінійні залежності,
– стійкіший до викидів та асиметричних розподілів.

**Отже, використовуємо кореляцію Спірмена для аналізу зв'язку між tip та total_bill у цьому випадку.**

**Додатково перевіримо через тести (D’Agostino-Pearson (normaltest) і Shapiro-Wilk) чи розподіли Tips та Total_bill є нормальними.**

```python
from scipy.stats import normaltest, shapiro


# tips["total_bill"]    tips['tip']

# Normaltest
_, p_value_normaltest_tips = normaltest(tips['tip'])
_, p_value_normaltest_totall_bill = normaltest(tips["total_bill"])

# Shapiro-Wilk test
_, p_value_shapiro_tips = shapiro(tips['tip'])
_, p_value_shapiro_total_bill = shapiro(tips["total_bill"])

# Test results
print(f"Normaltest for Tips data p-value: {p_value_normaltest_tips}")
print(f"Normaltest for Total_bill data p-value: {p_value_normaltest_totall_bill}")

print(f"Shapiro-Wilk for Tips data p-value: {p_value_shapiro_tips}")
print(f"Shapiro-Wilk for Total_bill data p-value: {p_value_shapiro_total_bill}")
```

Це результати тестів на нормальність для двох змінних: tip та total_bill. Проведено два статистичні тести: D’Agostino-Pearson (normaltest) і Shapiro-Wilk.

Normaltest for Tips data p-value: 5.796294322907102e-18\
Normaltest for Total_bill data p-value: 1.5951078766352608e-10\
Shapiro-Wilk for Tips data p-value: 8.200360128656659e-12\
Shapiro-Wilk for Total_bill data p-value: 3.3245391868090786e-10\

**Жоден із двох розподілів (tip, total_bill) не є нормально розподіленим. Обидва тести дали надзвичайно низькі p-value (набагато менші за 0.05), тому ми з високою впевненістю відхиляємо гіпотезу про нормальність.**

**Наслідки для кореляції:\
Оскільки дані не мають нормального розподілу, то краще використовувати коефіцієнт кореляції Спірмена.**

Крок 4. Розрахємо ступінь тісноти статистичного зв'язку між розміром групи size та розміром чека total_bill. Визначу, який із коефіцієнтів кореляції тут доречно застосувати та чи є зв'язок статистично значущим.

```python
from scipy.stats import pearsonr, spearmanr


# tips['size']    tips['total_bill']
# Pearson calculations

# Calculate statistic and p-value
statistic_and_p_value_pearson = pearsonr(tips['size'], tips['total_bill'])

# Calculate Pearson correlation
pearson_corr = tips["size"].corr(tips['total_bill'])

print(f"Pearson Test statistic and p-value: {statistic_and_p_value_pearson}")
print(f"Pearson Correlation: {pearson_corr}")



# Spearman calculations

# Calculate statistic and p-value
statistic_and_p_value_spearman = spearmanr(tips['size'], tips['total_bill'])

# Calculate Spearman correlation
spearman_corr = tips['tip'].corr(tips['total_bill'], method="spearman")

print(f"Spearman Test statistic and p-value: {statistic_and_p_value_spearman}")
print(f"Spearman Correlation: {spearman_corr}")
```

```python
# Перевірка size на нормальність
# не нормальний розподіл p < 0.05


# from scipy.stats import normaltest, shapiro

# D’Agostino and Pearson’s test
_, p_value_normaltest_size = normaltest(tips['size'])

# Shapiro-Wilk test
_, p_value_shapiro_size = shapiro(tips['size'])

print(f"Normaltest for size p-value: {p_value_normaltest_size}")
print(f"Shapiro-Wilk for size p-value: {p_value_shapiro_size}")
```

**Висновок щодо статистичного зв’язку між розміром групи та розміром чека**

З метою оцінки ступеня тісноти статистичного зв’язку між кількістю осіб у групі (змінна size) та сумою чека (змінна total_bill) було розраховано коефіцієнти кореляції Пірсона та Спірмена.

Попередній аналіз нормальності розподілу змінних size та total_bill, проведений за допомогою тестів Шапіро-Уїлка та Д’Агостіно-Пірсона, показав значні відхилення від нормального розподілу (p-value < 0.05 для обох змінних). У зв’язку з цим, основну увагу приділено інтерпретації коефіцієнта Спірмена, який не потребує нормальності.

**Коефіцієнт кореляції Спірмена становив 0.6048 при p-value ≈ 9.92×10⁻²⁶, що вказує на статистично значущий, середнього ступеня, монотонний прямий зв’язок між розміром групи та сумою чека.**

Для порівняння, коефіцієнт Пірсона становив 0.5983 при p-value ≈ 4.39×10⁻²⁵, що також вказує на статистично значущий прямий зв’язок, хоча за умов відсутності нормальності його використання є менш коректним.

**Отже, результати підтверджують наявність істотного позитивного зв’язку: зі збільшенням кількості осіб у групі, як правило, зростає і загальна сума чека.**
